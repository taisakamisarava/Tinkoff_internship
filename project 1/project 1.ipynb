```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

Я извлекла данные из базы в quests_analysis.csv при помощи SQL-запроса. А обработку и визуализацию сделала в Jupyter Notebook

```python
df = pd.read_csv('quests_analysis.csv', parse_dates = ['visit_dttm'])
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df = df.groupby([pd.Grouper(key = 'visit_dttm', freq = 'W')]).agg({
    'client_rk': 'count',
    'registration_flag': sum,
    'application_flag': sum,
    'finish_flag': sum
}).copy()
```

```python
df.head()
```

Отлично! Рассчитаем конверсию в регистрацию, в подачу заявки на игру, в приход на игру

```python
df['reg_conversion'] = df.registration_flag / df.client_rk
df['app_conversion'] = df.application_flag / df.client_rk
df['game_conversion'] = df.finish_flag / df.client_rk
```

```python
df.head()
```

```python
sns.set_palette("husl")
picture, (ax1, ax2) = plt.subplots(nrows = 2, figsize=(15,9))

# абсолютное число client_rk по датам соответственно
sns.lineplot(data = df, x = 'visit_dttm', y = 'client_rk', ax = ax1, color = 'orange')
ax1.set_ylabel('Количество клиентов')

# конверсии
sns.lineplot(data = df, x = 'visit_dttm', y = 'reg_conversion', ax = ax2)
sns.lineplot(data = df, x = 'visit_dttm', y = 'app_conversion', ax = ax2)
sns.lineplot(data = df, x = 'visit_dttm', y = 'game_conversion', ax = ax2)

ax2.set_ylabel('Конверсия')
ax2.xaxis.set_tick_params(which = 'both', labeltop = True)
ax2.legend([
    'Конверсия в регистрацию', 
    'Конверсия в подачу заявки', 
    'Конверсия в прохождение игры', ])
plt.subplots_adjust(hspace=.1)
plt.show()
```
